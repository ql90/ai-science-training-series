```python
import numpy as np
import matplotlib.pyplot as plt

for scale in [1,2,4,8]:
    # Load data from .dat file
    filename = 'metrics_' + str(scale) + '.dat'
    data = np.loadtxt(filename)
    # print(data)

    # scale = filename[-5]

    # Extract data columns
    train_acc = data[:, 0]
    train_loss = data[:, 1]
    valid_acc = data[:, 2]
    valid_loss = data[:, 3]
    time_per_epochs = data[:, 4]

    # Define the epochs
    epochs = np.arange(1, len(train_acc) + 1)

    # Plot training and validation accuracy curves
    plt.figure(figsize=(10, 6))
    plt.plot(epochs, train_acc, label='Training Accuracy')
    plt.plot(epochs, valid_acc, label='Validation Accuracy')
    plt.title(f'Training and Validation Accuracy Curves with {scale} nodes')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)
    plt.savefig(f'accuracy_{scale}.pdf')
    plt.show()
```
